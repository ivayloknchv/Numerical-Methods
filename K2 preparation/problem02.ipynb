```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def parabola(coeffs, x):
    a, b, c = coeffs[0], coeffs[1], coeffs[2]
    return a + b*x + c*x**2
```

```python
def create_row(x, n, pow):
    row = []
    for i in range (0, n+1):
        element = sum(x_node ** pow for x_node in x)
        row.append(element)
        pow+=1
    return row
```

```python
def create_matrix(x, n):
    matrix = []
    for i in range(0, n+1):
        matrix.append(create_row(x, n, i))
    return np.array(matrix)
```

```python
def create_vector(x, y, n):
    vector = []
    for i in range(0, n+1):
        curr = sum(y_node * (x_node**i) for (x_node, y_node) in zip(x,y))
        vector.append(curr)
    return np.array(vector)
```

```python
def least_squares_matrix(x, y, n):
    matrix = create_matrix(x, n)
    vector = create_vector(x, y, n)
    sol = np.linalg.solve(matrix, vector)
    return sol
```

```python
nodes = np.array([0.2, 0.5, 0.8, 1.2, 1.7, 2, 2.3])
```

```python
values = np.array([500, 700, 1000, 1200, 2200, 2650, 3750])
```

```python
coeffs = least_squares_matrix(nodes, values, 2)
```

```python
x_axis = np.linspace(min(nodes), max(nodes))
```

```python
y_axis = parabola(coeffs, x_axis)
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, y_axis)
plt.show()
```
