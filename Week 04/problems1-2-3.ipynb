```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def divided_difference(nodes, values):
    if len(nodes) == 1:
        return values[0]
    return (divided_difference(nodes[1::], values[1::]) - divided_difference(nodes[0:-1:], values[0:-1:])) / (nodes[len(nodes)-1] - nodes[0])                                                                                                  
```

```python
def newton_poly(nodes, values, x):
    res = 0
    multiplier = 1
    n = len(nodes)
    
    for k in range(0, n):
        current = divided_difference(nodes[0:k+1], values[0:k+1])*multiplier
        res+=current
        multiplier*= x-nodes[k]
        
    return res     
```

```python
def usa_population(x):
    years = np.arange(1920,2000, 10)
    population = np.array([106.46, 123.08, 132.12, 152.27, 180.67, 205.05, 227.23, 249.46])

    return newton_poly(years, population, x)
```

```python
x_axis = np.linspace(1920, 2000, 1000)
```

```python
plt.plot(x_axis, usa_population(x_axis))
```
