```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def calc_basis_func(nodes, k, x):
    res = 1
    for idx, node in enumerate(nodes):
        if idx != k:
            res *= np.sin((x-node)/2) / np.sin((nodes[k]-node)/2)
    return res
```

```python
def calc_trig_poly(nodes, values, x):
    res = 0
    for i in range(len(values)):
        res = values[i] * calc_basis_func(nodes, i, x)
    return res
```

```python
nodes = np.array([2, 5, 7, 11, 14,	16,	18,	21,	24])
values = np.array([7, 7.1, 7.4, 8.2, 8.9, 8.4, 7.9, 7.3, 7])
PERIOD = 24
transformed_nodes = np.array([(2 * np.pi * node) / PERIOD for node in nodes])
```

```python
x_axis = np.linspace(min(transformed_nodes), max(transformed_nodes), 1000)
```

```python
y_axis = calc_trig_poly(transformed_nodes, values, x_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```
