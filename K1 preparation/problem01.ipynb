```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def calc_basis_poly(nodes, k, x):
    res = 1
    for i in range(len(nodes)):
        if(i!=k):
            res*=(x-nodes[i])/(nodes[k]-nodes[i])
    return res
```

```python
def lagrange_poly(nodes, values, x):
    res = 0
    for i in range(len(nodes)):
        res += values[i] * calc_basis_poly(nodes, i, x)
    return res
```

```python
def abs_error(exact, appr):
    return abs(exact-appr)
```

```python
def relative_error(exact, appr):
    return abs_error(exact, appr) / exact
```

```python
nodes = np.array([0, 1.525, 3.05, 4.575, 6.1])
```

```python
values = np.array([1, 0.8617, 0.7385, 0.6292, 0.5328])
```

```python
x_axis = np.linspace(0, 6.1, 10000)
```

```python
y_axis = lagrange_poly(nodes, values, x_axis)
```

```python
plt.scatter(nodes, values)
plt.plot(x_axis, y_axis)
plt.show()
```

```python
sin_nodes = np.array([0, np.pi/6, np.pi/3, np.pi/2])
sin_values = np.array([np.sin(x) for x in sin_nodes])
```

```python
x_axis_sin = np.linspace(0, np.pi/2, 100)
```

```python
y_axis_exact = np.sin(x_axis_sin)
```

```python
y_axis_lagrange = lagrange_poly(sin_nodes, sin_values, x_axis_sin)
```

```python
plt.plot(x_axis_sin, y_axis_exact)
```

```python
plt.plot(x_axis_sin, y_axis_lagrange)
```

```python
plt.plot(x_axis_sin, abs(np.sin(x_axis_sin) - lagrange_poly(nodes, values, x_axis_sin)))
```
